```python
from sklearn.ensemble import RandomForestRegressor
import time

# Configurations for Random Forest testing
configs_rf = [
    {'id': 1,  'n_estimators': 50,  'max_depth': None, 'min_samples_split': 2, 'min_samples_leaf': 1},
    {'id': 2,  'n_estimators': 100, 'max_depth': None, 'min_samples_split': 2, 'min_samples_leaf': 1},
    {'id': 3,  'n_estimators': 50,  'max_depth': 10,   'min_samples_split': 5, 'min_samples_leaf': 2},
    {'id': 4,  'n_estimators': 100, 'max_depth': 10,   'min_samples_split': 5, 'min_samples_leaf': 2},
    {'id': 5,  'n_estimators': 100, 'max_depth': 20,   'min_samples_split': 2, 'min_samples_leaf': 1},
    {'id': 6,  'n_estimators': 200, 'max_depth': None, 'min_samples_split': 2, 'min_samples_leaf': 1},
    {'id': 7,  'n_estimators': 50,  'max_depth': 5,    'min_samples_split': 10,'min_samples_leaf': 4},
    {'id': 8,  'n_estimators': 100, 'max_depth': None, 'min_samples_split': 10,'min_samples_leaf': 2},
    {'id': 9,  'n_estimators': 300, 'max_depth': None, 'min_samples_split': 5, 'min_samples_leaf': 2},
    {'id': 10, 'n_estimators': 500, 'max_depth': 20,   'min_samples_split': 2, 'min_samples_leaf': 1},
    {'id': 11, 'n_estimators': 100, 'max_depth': 8,    'min_samples_split': 20, 'min_samples_leaf': 10},
    {'id': 12, 'n_estimators': 200, 'max_depth': 12,   'min_samples_split': 15, 'min_samples_leaf': 5},
    {'id': 13, 'n_estimators': 150, 'max_depth': 4,    'min_samples_split': 10, 'min_samples_leaf': 4},
    {'id': 14, 'n_estimators': 250, 'max_depth': None, 'min_samples_split': 50, 'min_samples_leaf': 5},
    {'id': 15, 'n_estimators': 150, 'max_depth': 25,   'min_samples_split': 4,  'min_samples_leaf': 2},
    {'id': 16, 'n_estimators': 100, 'max_depth': 15,   'min_samples_split': 8,  'min_samples_leaf': 3},
]

results_rf = []

print(f"--- Starting Random Forest tests ({len(configs_rf)} configurations) ---")

for config in configs_rf:
    print(f"Training RF ID {config['id']}...", end=" ")
    start_time = time.time()

    params = {k: v for k, v in config.items() if k != 'id'}

    # n_jobs=-1 uses all CPU cores for faster processing
    model = RandomForestRegressor(random_state=42, n_jobs=-1, **params)

    try:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        r2 = r2_score(y_test, y_pred)
        duration = time.time() - start_time

        print(f"Completed in {duration:.1f}s | RMSE: {rmse:.4f} | R2: {r2:.4f}")

        res = config.copy()
        res['RMSE'] = rmse
        res['R2'] = r2
        res['Time(s)'] = duration
        results_rf.append(res)

    except Exception as e:
        print(f"Error: {e}")

# Create DataFrame for analysis
df_results_rf = pd.DataFrame(results_rf)

# Show top results
print("\nTop 5 Best Random Forest Configurations:")
print(df_results_rf.sort_values(by='RMSE').head(5)[['id', 'RMSE', 'R2', 'n_estimators', 'max_depth']])
```

# Prediction Function

```python
import pandas as pd
import numpy as np

def make_prediction(new_data, model, scaler_for_model, feature_scaler):
  """
  Takes a dictionary of new data, processes it, and
  returns prediction in kW
  """
  scaler_input = pd.DataFrame({
      'Global_active_power': [0],
      'Global_reactive_power': [new_data['Global_reactive_power']],
      'Voltage': [new_data['Voltage']],
      'Global_intensity': [0],
      'Sub_metering_1': [new_data['Sub_metering_1']],
      'Sub_metering_2': [new_data['Sub_metering_2']],
      'Sub_metering_3': [new_data['Sub_metering_3']]
  })

  normalized_data = feature_scaler.transform(scaler_input)
  cols = ['Global_active_power', 'Global_reactive_power', 'Voltage', 'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']
  df_norm = pd.DataFrame(normalized_data, columns=cols)

  datetime_val = pd.to_datetime(new_data['data_hora'])

  X_input_df = pd.DataFrame({
      'Global_reactive_power': df_norm['Global_reactive_power'],
      'Voltage': df_norm['Voltage'],
      'Sub_metering_1': df_norm['Sub_metering_1'],
      'Sub_metering_2': df_norm['Sub_metering_2'],
      'Sub_metering_3': df_norm['Sub_metering_3'],
      'hour': [datetime_val.hour],
      'day_of_week': [datetime_val.dayofweek],
      'month': [datetime_val.month],
      'year': [datetime_val.year]
  })


  X_input_scaled = scaler_for_model.transform(X_input_df)

  prediction_scaled = model.predict(X_input_scaled)[0]


  min_target = feature_scaler.data_min_[0]
  max_target = feature_scaler.data_max_[0]

  prediction_final = prediction_scaled * (max_target - min_target) + min_target

  return prediction_final

new_reading = {
    'data_hora': '2024-12-16 17:28:00',
    'Global_reactive_power': 0.528,
    'Voltage': 235.68,
    'Sub_metering_1': 0.0,
    'Sub_metering_2': 1.0,
    'Sub_metering_3': 17.0
}


predicted_consumption = make_prediction(new_reading, MLP, scaler_MLP, scaler)

print("--- ENERGY CONSUMPTION PREDICTION ---")
print(f"Date/Time: {new_reading['data_hora']}")
print(f"Predicted Consumption: {predicted_consumption:.4f} kW")
```

# Policy Simulation

```python
df_simulated_policy = df_ready.copy()

# Apply the percentage reductions to the Sub_metering columns
# Sub_metering_1: 10% reduction
df_simulated_policy['Sub_metering_1'] = df_simulated_policy['Sub_metering_1'] * (1 - 0.10)

# Sub_metering_2: 15% reduction
df_simulated_policy['Sub_metering_2'] = df_simulated_policy['Sub_metering_2'] * (1 - 0.15)

# Sub_metering_3: 5% reduction
df_simulated_policy['Sub_metering_3'] = df_simulated_policy['Sub_metering_3'] * (1 - 0.05)

# Recalculate Global_active_power if it's derived from sub-meterings, otherwise consider it as affected separately.
# In this dataset, Global_active_power is a primary measurement, so we'll assume a direct impact.
# A general reduction in Global_active_power is expected, let's estimate an overall 5% reduction based on the sub-metering impacts.
# This is an assumption for simulation purposes, actual impact might require a more complex model.
df_simulated_policy['Global_active_power'] = df_simulated_policy['Global_active_power'] * (1 - 0.05)

# Display the first few rows of the simulated policy DataFrame to show the changes
print("First 5 rows of df_simulated_policy:")
print(df_simulated_policy.head())

# Display descriptive statistics for the affected columns to confirm changes
print("\nDescriptive statistics for affected columns in df_simulated_policy:")
print(df_simulated_policy[['Global_active_power', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].describe())
```

```python
def simulate_policy(df_base, scenario_config, X_test_index):
    """
    Applies the reductions specified in a policy scenario to a DataFrame
    and returns the modified DataFrame, focusing only on the `X_test` portion.
    """
    df_sim = df_base.copy()

    # Apply reductions
    df_sim['Sub_metering_1'] = df_sim['Sub_metering_1'] * (1 - scenario_config['sub_1_reduction'])
    df_sim['Sub_metering_2'] = df_sim['Sub_metering_2'] * (1 - scenario_config['sub_2_reduction'])
    df_sim['Sub_metering_3'] = df_sim['Sub_metering_3'] * (1 - scenario_config['sub_3_reduction'])
    df_sim['Global_active_power'] = df_sim['Global_active_power'] * (1 - scenario_config['global_active_power_reduction'])

    # Select only the relevant feature columns, aligned with X_test index
    # and ensure the order of columns matches X_train/X_test
    feature_cols = X_train.columns.tolist() # Assuming X_train has the correct order of features
    X_simulated_policy = df_sim.loc[X_test_index, feature_cols]

    return X_simulated_policy

all_scenario_predictions = {}

print("\n--- Simulating and Predicting for Each Policy Scenario ---")

for scenario in policy_scenarios:
    scenario_name = scenario['name']
    print(f"Simulating '{scenario_name}'...")

    # 1. Simulate the policy impact on the features
    X_sim_scenario = simulate_policy(df_ready, scenario, X_test.index)

    # 2. Scale the simulated features using the MLP scaler
    X_sim_scenario_scaled = scaler_MLP.transform(X_sim_scenario)

    # 3. Make predictions using the MLP model
    pred_scenario = MLP.predict(X_sim_scenario_scaled)

    all_scenario_predictions[scenario_name] = pred_scenario

    # Calculate and print metrics for each scenario
    rmse_scenario = np.sqrt(mean_squared_error(y_test, pred_scenario)) # Compare against actual y_test
    r2_scenario = r2_score(y_test, pred_scenario)

    mean_baseline = MLP_pred.mean()
    mean_policy = pred_scenario.mean()
    percentage_reduction = ((mean_baseline - mean_policy) / mean_baseline) * 100

    print(f"  RMSE: {rmse_scenario:.4f}, R2: {r2_scenario:.4f}")
    print(f"  Mean Baseline Prediction: {mean_baseline:.4f}")
    print(f"  Mean Scenario Prediction: {mean_policy:.4f}")
    print(f"  Percentage Reduction in Mean Energy Consumption: {percentage_reduction:.2f}%\n")


print("\n--- Policy Simulation Analysis Complete ---")
```

```python
import gradio as gr
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
import cohere
from google.colab import userdata
import os

# --- 1. Load Pre-trained Model and Scalers ---
# Ensure these files are in the same directory as this script
model_filename = 'mlp_regressor_model.joblib'
scheduler_filename = 'scaler_mlp.joblib'
feature_scaler_filename = 'feature_scaler.joblib'

try:
    MLP = joblib.load(model_filename)
    scaler_MLP = joblib.load(scheduler_filename)
    scaler = joblib.load(feature_scaler_filename)
    print("Models and scalers loaded successfully.")
except FileNotFoundError:
    print(f"Error: One or more files not found: {model_filename}, {scheduler_filename}, {feature_scaler_filename}")
    print("Please ensure the model and scaler files are saved in the current directory.")
    MLP, scaler_MLP, scaler = None, None, None

# Assuming these variables are available from the notebook's global scope
# (X_train, X_test, y_test, MLP_pred, all_scenario_predictions, policy_scenarios, policy_description, qualitative_insights, df_ready)
# If running standalone, these would need to be loaded/defined.

# --- 2. Prediction Function ---
def make_prediction(data_hora, global_reactive_power, voltage, sub_metering_1, sub_metering_2, sub_metering_3):
    if MLP is None or scaler_MLP is None or scaler is None:
        return "Error: Model or scalers not loaded."

    new_data = {
        'data_hora': data_hora,
        'Global_reactive_power': global_reactive_power,
        'Voltage': voltage,
        'Sub_metering_1': sub_metering_1,
        'Sub_metering_2': sub_metering_2,
        'Sub_metering_3': sub_metering_3
    }

    scaler_input = pd.DataFrame({
        'Global_active_power': [0],
        'Global_reactive_power': [new_data['Global_reactive_power']],
        'Voltage': [new_data['Voltage']],
        'Global_intensity': [0],
        'Sub_metering_1': [new_data['Sub_metering_1']],
        'Sub_metering_2': [new_data['Sub_metering_2']],
        'Sub_metering_3': [new_data['Sub_metering_3']]
    })

    cols = ['Global_active_power', 'Global_reactive_power', 'Voltage', 'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']
    normalized_data = scaler.transform(scaler_input[cols])
    df_norm = pd.DataFrame(normalized_data, columns=cols)

    datetime_val = pd.to_datetime(new_data['data_hora'])

    # Ensure the order of columns matches X_train
    feature_cols_order = X_train.columns.tolist() if 'X_train' in globals() else ['Global_reactive_power', 'Voltage', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3', 'hour', 'day_of_week', 'month', 'year']

    X_input_df = pd.DataFrame({
        'Global_reactive_power': df_norm['Global_reactive_power'],
        'Voltage': df_norm['Voltage'],
        'Sub_metering_1': df_norm['Sub_metering_1'],
        'Sub_metering_2': df_norm['Sub_metering_2'],
        'Sub_metering_3': df_norm['Sub_metering_3'],
        'hour': [datetime_val.hour],
        'day_of_week': [datetime_val.dayofweek],
        'month': [datetime_val.month],
        'year': [datetime_val.year]
    }, columns=feature_cols_order)

    X_input_scaled = scaler_MLP.transform(X_input_df)
    prediction_scaled = MLP.predict(X_input_scaled)[0]

    min_target = scaler.data_min_[0]
    max_target = scaler.data_max_[0]

    prediction_final = prediction_scaled * (max_target - min_target) + min_target

    return f"{prediction_final:.4f} kW"

# --- 3. Policy Simulation Function ---
def simulate_policy_gradio(policy_name):
    global df_ready, X_test, y_test, MLP_pred, policy_scenarios, X_train # X_train needed for feature_cols

    if MLP is None or scaler_MLP is None or scaler is None or df_ready is None or X_test is None or y_test is None or MLP_pred is None or policy_scenarios is None or X_train is None:
        return gr.Markdown("Error: Dependencies (model, scalers, or data) not loaded. Please run previous cells."), None, None

    selected_scenario = next((s for s in policy_scenarios if s['name'] == policy_name), None)
    if not selected_scenario:
        return gr.Markdown(f"Error: Policy scenario '{policy_name}' not found."), None, None

    df_sim = df_ready.copy()

    df_sim['Sub_metering_1'] = df_sim['Sub_metering_1'] * (1 - selected_scenario['sub_1_reduction'])
    df_sim['Sub_metering_2'] = df_sim['Sub_metering_2'] * (1 - selected_scenario['sub_2_reduction'])
    df_sim['Sub_metering_3'] = df_sim['Sub_metering_3'] * (1 - selected_scenario['sub_3_reduction'])
    df_sim['Global_active_power'] = df_sim['Global_active_power'] * (1 - selected_scenario['global_active_power_reduction'])

    feature_cols = X_train.columns.tolist()
    X_simulated_policy = df_sim.loc[X_test.index, feature_cols]

    X_simulated_policy_scaled = scaler_MLP.transform(X_simulated_policy)
    MLP_policy_pred = MLP.predict(X_simulated_policy_scaled)

    df_comparison = pd.DataFrame({
        'Baseline_Prediction': MLP_pred,
        'Policy_Prediction': MLP_policy_pred
    }, index=y_test.index)

    mean_baseline = df_comparison['Baseline_Prediction'].mean()
    mean_policy = df_comparison['Policy_Prediction'].mean()
    percentage_reduction = ((mean_baseline - mean_policy) / mean_baseline) * 100

    fig, ax = plt.subplots(figsize=(12, 6))
    plot_segment_length = min(len(y_test), 200)
    ax.plot(y_test.index[:plot_segment_length], y_test[:plot_segment_length], label='Actual', color='green', alpha=0.7)
    ax.plot(df_comparison.index[:plot_segment_length], df_comparison['Baseline_Prediction'][:plot_segment_length], label='Baseline Prediction', color='blue', linestyle='--')
    ax.plot(df_comparison.index[:plot_segment_length], df_comparison['Policy_Prediction'][:plot_segment_length], label=f'{policy_name} Prediction', color='red', linestyle='-.')
    ax.set_title(f'Time Series Comparison: Actual, Baseline, and {policy_name} Predictions (First {plot_segment_length} Hours)')
    ax.set_xlabel('Time')
    ax.set_ylabel('Global Active Power (kW, scaled)')
    ax.legend()
    plt.tight_layout()
    plt.close(fig) # Close the figure to free up memory

    summary_text = (
        f"### 📊 Policy Simulation Results for '{policy_name}'\n"
        f"- Mean Baseline Prediction (scaled): {mean_baseline:.4f}\n"
        f"- Mean {policy_name} Prediction (scaled): {mean_policy:.4f}\n"
        f"- Percentage Reduction: <span style='color:green;'>{percentage_reduction:.2f}%</span>"
    )

    return summary_text, fig, MLP_policy_pred

# --- 4. Scenario Comparison Function ---
def scenario_comparison_gradio():
    global y_test, MLP_pred, all_scenario_predictions, policy_scenarios

    if MLP is None or scaler_MLP is None or scaler is None or y_test is None or MLP_pred is None or all_scenario_predictions is None or policy_scenarios is None:
        return None # Return None when error, since output is gr.Plot

    fig, ax = plt.subplots(figsize=(15, 8))
    plot_segment_length = min(len(y_test), 300)

    ax.plot(y_test.index[:plot_segment_length], y_test[:plot_segment_length], label='Actual', color='gray', alpha=0.8, linewidth=2)
    ax.plot(y_test.index[:plot_segment_length], MLP_pred[:plot_segment_length], label='Baseline Prediction', color='blue', linestyle='--', alpha=0.7)

    colors = plt.cm.viridis(np.linspace(0, 1, len(policy_scenarios)))

    for i, scenario in enumerate(policy_scenarios):
        scenario_name = scenario['name']
        if scenario_name in all_scenario_predictions:
            predictions = all_scenario_predictions[scenario_name]
            ax.plot(y_test.index[:plot_segment_length], predictions[:plot_segment_length], label=f'{scenario_name} Prediction', linestyle=':', alpha=0.7, color=colors[i])

    ax.set_title(f'Time Series Comparison: Actual, Baseline, and All Policy Scenarios (First {plot_segment_length} Hours)', fontsize=16)
    ax.set_xlabel('Time')
    ax.set_ylabel('Global Active Power (kW, scaled)')
    ax.legend()
    plt.tight_layout()
    plt.close(fig) # Close the figure to free up memory
    return fig

# --- 5. LLM Analysis Function ---
def generate_llm_analysis(policy_sim_output_str, policy_pred_array):
    global policy_description, qualitative_insights, MLP_pred, all_scenario_predictions, policy_scenarios

    cohere_api_key = userdata.get('COHERE_API_KEY')
    if not cohere_api_key:
        return "Cohere API key not found. Please set 'COHERE_API_KEY' in Colab secrets."

    if MLP is None or scaler_MLP is None or scaler is None or MLP_pred is None or all_scenario_predictions is None or policy_description is None or qualitative_insights is None:
        return "Error: Dependencies (model, scalers, or data) not loaded, cannot provide LLM analysis. Please run previous cells."

    co = cohere.Client(cohere_api_key)

    quantitative_results = "\n"
    quantitative_results += f"Mean Baseline Prediction (scaled): {MLP_pred.mean():.4f}\n"
    for scenario_name, predictions in all_scenario_predictions.items():
        mean_scenario_pred = predictions.mean()
        percentage_reduction = ((MLP_pred.mean() - mean_scenario_pred) / MLP_pred.mean()) * 100
        quantitative_results += f"Mean {scenario_name} Prediction (scaled): {mean_scenario_pred:.4f}\n"
        quantitative_results += f"Percentage Reduction for {scenario_name}: {percentage_reduction:.2f}%\n"

    llm_objective = """
    Your objective is to:
    1. Generate an enriched summary of the policy's simulated impact across all scenarios.
    2. Formulate specific policy recommendations based on the comparative findings.
    3. Explore potential alternative scenarios or unintended consequences of implementing such policies, considering their varying impacts.
    """

    llm_prompt_for_gradio = f"""
    Based on the following energy efficiency policy simulation:

    ### Policy Description:
    {policy_description}

    ### Quantitative Simulation Results:
    {quantitative_results}

    ### Qualitative Insights from Analysis:
    {qualitative_insights}

    ### Your Task:
    {llm_objective}
    """

    try:
        response = co.chat(
            model='command-a-03-2025',
            message=llm_prompt_for_gradio,
            temperature=0.7
        )
        return response.text
    except Exception as e:
        return f"Error calling Cohere API: {e}"


# --- Gradio UI Layout ---
with gr.Blocks(theme=gr.themes.Monochrome()) as demo:
    gr.Markdown(
        """<center><h1>⚡ Smart Energy Policy Simulator ⚡</h1></center>"""
        """<center>Predict household energy consumption and simulate policy impacts.</center>"""
    )


    with gr.Tabs() as tabs:
        with gr.Tab(label="Individual Prediction 💡"):
            gr.Markdown("## Section 1: Input Panel for Energy Consumption Prediction")
            with gr.Row():
                datetime_input = gr.Textbox(label="Date and Time (YYYY-MM-DD HH:MM:SS)", placeholder="e.g., 2006-12-16 17:28:00", value="2006-12-16 17:28:00")
            with gr.Row():
                reactive_power_input = gr.Slider(minimum=0, maximum=1.5, value=0.528, label="Global Reactive Power", info="(0.0-1.5)")
                voltage_input = gr.Slider(minimum=220, maximum=260, value=235.68, label="Voltage", info="(220-260)")
            with gr.Row():
                sub_metering_1_input = gr.Slider(minimum=0, maximum=90, value=0.0, label="Sub Metering 1 (Kitchen/Laundry)", info="(0-90 kW)")
                sub_metering_2_input = gr.Slider(minimum=0, maximum=80, value=1.0, label="Sub Metering 2 (Heating/Lighting)", info="(0-80 kW)")
                sub_metering_3_input = gr.Slider(minimum=0, maximum=40, value=17.0, label="Sub Metering 3 (Other Appliances)", info="(0-40 kW)")

            predict_btn = gr.Button("Predict Energy Consumption ✨")
            gr.Markdown("## Section 2: Prediction Output")
            prediction_output = gr.Textbox(label="Predicted Global Active Power", interactive=False)

            predict_btn.click(
                fn=make_prediction,
                inputs=[datetime_input, reactive_power_input, voltage_input, sub_metering_1_input, sub_metering_2_input, sub_metering_3_input],
                outputs=prediction_output
            )

        with gr.Tab(label="Policy Simulation 📈"):
            gr.Markdown("## Section 3: Policy Simulation")
            policy_choice = gr.Dropdown(
                label="Select Policy Scenario",
                choices=[s['name'] for s in policy_scenarios] if 'policy_scenarios' in globals() else [],
                value="Original Policy" if 'policy_scenarios' in globals() else None
            )
            simulate_policy_btn = gr.Button("Simulate Policy Impact 🚀")

            policy_simulation_text = gr.Markdown("", label="Policy Simulation Summary")
            policy_simulation_plot = gr.Plot(label="Baseline vs. Policy Predictions")
            policy_simulated_predictions = gr.State(value=None)

            simulate_policy_btn.click(
                fn=simulate_policy_gradio,
                inputs=policy_choice,
                outputs=[policy_simulation_text, policy_simulation_plot, policy_simulated_predictions]
            )

        with gr.Tab(label="Scenario Comparison 📊"):
            gr.Markdown("## Section 4: Scenario Comparison")
            scenario_plot = gr.Plot(label="All Policy Scenarios vs. Baseline")
            compare_scenarios_btn = gr.Button("Compare All Scenarios ⚖️")

            compare_scenarios_btn.click(
                fn=scenario_comparison_gradio,
                inputs=[],
                outputs=scenario_plot
            )

        with gr.Tab(label="LLM Analysis 🧠"):
            gr.Markdown("## Section 5: LLM-Powered Policy Analysis")
            llm_output_text = gr.Markdown(
                "Click the button to generate an in-depth analysis and recommendations based on the policy simulations. (This may take a moment)."
            )
            generate_llm_btn = gr.Button("Generate LLM Analysis 🦾")

            generate_llm_btn.click(
                fn=generate_llm_analysis,
                inputs=[policy_simulation_text, policy_simulated_predictions], # These inputs are used for context, but the function primarily uses globals
                outputs=llm_output_text
            )

# --- Launch the Gradio App ---
if __name__ == "__main__":
    demo.launch(debug=True)
```

### Project Summary

This project focused on analyzing and predicting household power consumption, and simulating the impact of energy efficiency policies. Here's a breakdown of the process:

### 1. Data Loading and Initial Exploration
- The project started by loading the `household_power_consumption.txt` dataset into a pandas DataFrame. Initial checks revealed the dataset contains over 2 million entries with 9 columns, including `Date`, `Time`, and various power consumption metrics.
- Data types were inspected, showing `Date` and `Time` as objects, and other power metrics as `float64`. Missing values were identified across all numerical power columns, with each having 25,979 null entries.

### 2. Data Preprocessing
- **Time Series Adjustment**: The `Date` and `Time` columns were combined and converted into a single `datetime` index, making the DataFrame suitable for time series analysis. The original `Date` and `Time` columns were then dropped.
- **Null Handling**: Missing values were handled using time-based interpolation (`method='time'`), effectively filling all 25,979 nulls in each numerical column.
- **Resampling and Feature Engineering**: The data was resampled to an hourly frequency, and new time-based features (`hour`, `day_of_week`, `month`, `year`) were extracted from the datetime index. This reduced the dataset size from over 2 million to 34,589 hourly entries.
- **Normalization**: Key consumption features (`Global_active_power`, `Global_reactive_power`, `Voltage`, `Global_intensity`, `Sub_metering_1`, `Sub_metering_2`, `Sub_metering_3`) were scaled using `MinMaxScaler` to a range of (0,1), preparing them for machine learning models.

### 3. Model Training and Evaluation
- **Data Splitting**: The processed data was split into training and testing sets (80/20 split) with `Global_active_power` as the target variable (`y`) and other relevant features as predictors (`X`). `Global_intensity` was dropped as it's highly correlated with `Global_active_power`.
- **Exploratory Data Analysis (EDA)**: Correlation analysis and visualizations (time series plots, box plots by hour and month) were performed to understand data distributions and relationships.
- **Model Selection and Performance**: Several regression models were trained and evaluated based on RMSE and R2 score:
    - **Linear Regression**: Achieved an RMSE of 0.0504 and R2 of 0.8016.
    - **MLP Regressor (Multi-Layer Perceptron)**: After scaling features with `StandardScaler`, the MLP model achieved the best performance with an RMSE of 0.0416 and R2 of 0.8647. This model was chosen for policy simulation.
    - **Random Forest Regressor**: Performed strongly with an RMSE of 0.0421 and R2 of 0.8616. Further hyperparameter tuning was explored.
    - **XGBoost Regressor**: Achieved an RMSE of 0.0428 and R2 of 0.8571.
- **Feature Importance**: SHAP (SHapley Additive exPlanations) values were used to analyze feature importance for the best-performing MLP model, providing insights into which features drive predictions.

### 4. Policy Simulation and Analysis
- An energy efficiency policy targeting low-income households was simulated, involving reductions in sub-metered consumption (e.g., 10% for Sub_metering_1, 15% for Sub_metering_2, 5% for Sub_metering_3) and an overall reduction in `Global_active_power`.
- Three policy scenarios were defined: 'Original Policy', 'Aggressive Policy', and 'Moderate Policy', each with different reduction percentages.
- The MLP model was used to predict energy consumption under each simulated policy. The results showed a percentage reduction in mean energy consumption across all scenarios:
    - **Original Policy**: 3.76% reduction
    - **Aggressive Policy**: 11.17% reduction (highest impact)
    - **Moderate Policy**: 1.83% reduction
- Visualizations (histograms, box plots, time series comparisons) demonstrated a clear downward shift in predicted consumption under policy scenarios compared to the baseline, indicating the effectiveness of the interventions.

### 5. LLM-Powered Policy Analysis
- The simulated results, along with a detailed policy description and qualitative insights, were fed into a Cohere LLM (Large Language Model) to generate an enriched summary, specific policy recommendations, and explore potential alternative scenarios or unintended consequences.

### 6. Gradio Interface

A user-friendly Gradio interface was developed (code in cell `tfdXSmFrkUVz`) to interact with the trained model and policy simulations. The interface consists of four main tabs:

*   **Individual Prediction 💡**: Allows users to input specific energy parameters (date/time, reactive power, voltage, sub-metering values) and receive an instant prediction of Global Active Power consumption in kW.

*   **Policy Simulation 📈**: Enables users to select one of the predefined policy scenarios ('Original Policy', 'Aggressive Policy', 'Moderate Policy'). Upon selection, it displays a summary of the simulated impact, including mean baseline and policy predictions, and the percentage reduction in mean energy consumption. A time series plot visualizes the actual consumption, baseline prediction, and the selected policy's prediction over the first 200 hours.

*   **Scenario Comparison 📊**: This tab provides a comparative visualization of all policy scenarios against the actual consumption and baseline prediction. A single time series plot illustrates the effect of each policy over the first 300 hours, allowing for easy comparison of their impacts.

*   **LLM Analysis 🧠**: By clicking a button, users can trigger an LLM (Cohere) to generate an in-depth analysis of the policy simulations. This includes an enriched summary of the policy's impact, specific policy recommendations, and an exploration of potential alternative scenarios or unintended consequences, leveraging the quantitative and qualitative insights from the project.
